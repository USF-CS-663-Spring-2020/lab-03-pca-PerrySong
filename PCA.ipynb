# WDBC Dataset PCA

```python
import pandas as pd
df = pd.read_csv('wdbc.data', header=None)
df.head()
```

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']

df.columns = column_labels
df['diagnosis'].replace('B', 'Benign', inplace=True)
df['diagnosis'].replace('M', 'Malignant', inplace=True)
df.head()
```

```python
feature_df = df.iloc[:, 2:]
label_df = df.iloc[:, 1]
# df.iloc[:,1].unique()
feature_df.head()
```

```python
label_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

x = feature_df.values
x = StandardScaler().fit_transform(x) # could also do StandardScaler().fit_transform(feature_df)
x
```

```python
import numpy as np

np.mean(x), np.std(x)
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2) # orthogona
principal_Components_breast = pca_breast.fit_transform(x)
principal_breast_df = pd.DataFrame(data=principal_Components_breast, \
                                   columns = ['principal component 1', 'principal component 2'])

pca_breast.explained_variance_ratio_ # % variance explained
```

```python
principal_breast_df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(10, 10))
plt.xlabel('PC 1')
plt.ylabel('PC 2')
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target, color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_breast_df.loc[indicesToKeep, 'principal component 1'], \
               principal_breast_df.loc[indicesToKeep, 'principal component 2'], \
               c = color)
```

```python
# How many principal components do we need to explain the variance in data?
pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
# FYI: We have to go to 6 significant digits to explain the last PC.
```

```python
# Graph it...
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```

# Gisette Dataset PCA

```python
df = pd.read_csv('gisette_train.data', header = None, delim_whitespace = True)
df
```

```python
label_df = pd.read_csv('gisette_train.labels', header = None).iloc[:, 0]
label_df.head()
```

```python
x = df.values
x = StandardScaler().fit_transform(x)
x
```

```python
np.mean(x), np.std(x)
```

```python
# For the first two features:
pca = PCA(n_components = 2)
principal_components = pca.fit_transform(x)
principal_df = pd.DataFrame(data=principal_components, \
                                   columns = ['principal component 1', 'principal component 2'])
pca.explained_variance_ratio_
```

```python
principal_df
```

```python
%matplotlib inline
plt.figure()
plt.figure(figsize=(10, 10))
plt.title("Gisette PCA",fontsize=25)
plt.xlabel('PC 1')
plt.ylabel('PC 2')

targets = [-1, 1]
colors = ['b', 'r']

for target, color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_df.loc[indicesToKeep, 'principal component 1'], \
               principal_df.loc[indicesToKeep, 'principal component 2'], \
               c = color)
```

```python
pca = PCA().fit(x)

evr = pca.explained_variance_ratio_

print('\n'.join('{:d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

```python
# Graph it...
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```

# Dorothea Dataset PCA

```python
df = pd.read_csv('dorothea_train.data', header=None, delim_whitespace=True, error_bad_lines=False)
df.head()
```

```python
d_labels = pd.read_csv('dorothea_train.labels').iloc[:,0]
d_labels.unique()
```

```python
x = df.fillna(0).values # Fill the Nan with 0
x = StandardScaler().fit_transform(x) # could also do StandardScaler().fit_transform(feature_df)
x
```

```python
np.mean(x), np.std(x)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)
principal_df = pd.DataFrame(data=principal_components, \
                                   columns = ['principal component 1', 'principal component 2'])
pca.explained_variance_ratio_
```

```python
# Scatter plot for the first two features PCA
plt.figure()
plt.figure(figsize=(10, 10))
plt.xlabel('PC 1')
plt.ylabel('PC 2')
targets = [-1, 1]
colors = ['b', 'r']
for target, color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_df.loc[indicesToKeep, 'principal component 1'], \
               principal_df.loc[indicesToKeep, 'principal component 2'], \
               c = color)
```

```python
# Calculate every principal component for the dataset
pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

```python
# Graph it...
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```
